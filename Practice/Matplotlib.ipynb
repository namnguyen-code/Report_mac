```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
b
```

```python
fig_1 = plt.figure()
axes = fig_1.add_axes([0,0,1,1])
axes.plot(a,b)
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a,b)

axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(x,y)

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a,b)
axes1.set_xlim(0, 8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('a')
axes1.set_ylabel('b')
axes1.set_title('Power of four')

axes2 = fig.add_axes([0.25, 0.25, 0.25, 0.25])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('a')
axes1.set_ylabel('b')
axes2.set_title('Zoomed In')

axes3 = fig.add_axes([1,1, 0.25, 0.25])
axes3.plot(a,b)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

fig.suptitle('Figure Level')

axes[0][0].plot(a,b)
axes[0][0].set_title('0 0 Title')

axes[0][1].plot(x,y)

axes[1][0].plot(b,a)

axes[1][1].plot(y,x)
axes[1][1].set_title('1 1 title')
axes[1][1].set_xlabel('1 1 X Label')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

fig.suptitle('Figure Level')

axes[0][0].plot(a,b)
axes[0][0].set_title('0 0 Title')

axes[0][1].plot(x,y)

axes[1][0].plot(b,a)

axes[1][1].plot(y,x)
axes[1][1].set_title('1 1 title')
axes[1][1].set_xlabel('1 1 X Label')

fig.subplots_adjust(
    left = 0.125,
    right = 0.9,
    bottom= 0.01,
    top = 0.9,
    wspace= 0.1,
    hspace = 0.1
)

plt.show()
```
